```sql
use AdventureWorks2019;


```

```sql
begin tran t1
    select ProductID, avg(LineTotal) avg_line_total
    into #tmp
    from Sales.SalesOrderDetail
    group by ProductID;


    select sod.ProductID, sum(sod.LineTotal) total_LineTotal,
        count(sod.SalesOrderID) num_orders,
        count(distinct sod.SalesOrderID) num_distinct_orders,
        sum(sod.OrderQty) total_items
    from Sales.SalesOrderDetail sod
        join #tmp c on sod.ProductID=c.ProductID and sod.LineTotal>avg_line_total
    group by sod.ProductID;

commit tran t1
```

```sql
select ProductID, 1 ProductGroup, avg(LineTotal) avg_line_total
into dbo.ProductAvg
from Sales.SalesOrderDetail
group by ProductID;

```

```sql
begin tran t1
select count(*) from dbo.ProductAvg;

insert dbo.ProductAvg
select ProductID, 2 ProductGroup, avg(LineTotal) avg_line_total
    from Sales.SalesOrderDetail
    group by ProductID;

select count(*) from dbo.ProductAvg;

rollback tran t1
select count(*) from dbo.ProductAvg;
drop table if exists dbo.ProductAvg;
```

```sql
drop table if exists Sales.SalesOrderDetail_copy;

select sod.*, p.Name ProductName, p.ProductNumber, p.Color ProductColor, p.SellStartDate
into Sales.SalesOrderDetail_copy
from Sales.SalesOrderDetail sod join Production.Product p on sod.ProductID=p.ProductId;

select count(*) from Sales.SalesOrderDetail_copy;
```
